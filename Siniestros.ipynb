```python
import pandas as pd
```

```python
url_siniestros = "https://raw.githubusercontent.com/gaby1719/etl-data-pipeline/refs/heads/main/data/raw/siniestros.csv"
```

```python
siniestros = pd.read_csv(url_siniestros)
```

```python
siniestros.info()
```

```python
def limpiar_dataframe(df):



    df.columns = df.columns.str.strip().str.lower()



    for col in df.select_dtypes(include='object').columns:

        df[col] = df[col].astype(str).str.strip()



    df = df.replace(r'^\s*$', pd.NA, regex=True)



    df = df.drop_duplicates()



    return df
```

```python
siniestros['fecha_siniestro'] = pd.to_datetime(

    siniestros['fecha_siniestro'], errors='coerce'

)
```

```python
siniestros.to_csv("siniestros_curated.csv", index=False)
```
