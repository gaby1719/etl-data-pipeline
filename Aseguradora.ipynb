```python
import pandas as pd
```

```python
url_aseguradoras = "https://raw.githubusercontent.com/gaby1719/etl-data-pipeline/refs/heads/main/data/raw/aseguradoras.csv"
```

```python
aseguradoras = pd.read_csv(url_aseguradoras)
```

```python
aseguradoras.info()
```

```python
def limpiar_dataframe(df):



    df.columns = df.columns.str.strip().str.lower()



    for col in df.select_dtypes(include='object').columns:

        df[col] = df[col].astype(str).str.strip()



    df = df.replace(r'^\s*$', pd.NA, regex=True)



    df = df.drop_duplicates()



    return df
```

```python
aseguradoras.to_csv("aseguradoras_curated.csv", index=False)
```
